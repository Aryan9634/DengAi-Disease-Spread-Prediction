Loading data

```python
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
    
def load_data(data_dir, file_name, date_columns=[]):
    """
    Load data from a CSV file and parse date columns.

    Parameters:
        data_dir (str): The directory where the data file is located.
        file_name (str): The name of the CSV file to load.
        date_columns (list): List of column names to be parsed as dates.

    Returns:
        pd.DataFrame: The loaded dataset.
    """
    file_path = os.path.join(data_dir, file_name)
    return pd.read_csv(file_path, parse_dates=date_columns)
```

```python
X_train = load_data(data_dir, "dengue_features_train.csv", date_columns=["week_start_date"])
y_train = load_data(data_dir, "dengue_labels_train.csv")["total_cases"]
data = pd.concat([X_train,y_train],axis=1)
X_test = load_data(data_dir, "dengue_features_test.csv", date_columns=["week_start_date"])
```

- Null value processing 

```python
def handle_null_values(data, threshold=10):
    """
    Handle null values in the dataset.

    Parameters:
        data (pd.DataFrame): The input dataset.
        threshold (float): The threshold percentage for null values in a feature.
        
    Returns:
        pd.DataFrame: The dataset with null values processed.
    """
    print("Null Value Handling:")
    
    # Calculate and display null value statistics
    total_null_values = data.isnull().sum().sum()
    null_percentages = (data.isnull().sum() / len(data)) * 100
    print(f"Total null values before processing: {total_null_values}")
    print("Feature-specific null percentages:")
    print(null_percentages)
    
    # Identify columns with null percentages exceeding the threshold
    columns_to_impute = [col for col, percentage in null_percentages.items() if percentage > threshold]
    
    # Impute identified columns with the mean value
    for col in columns_to_impute:
        data[col].fillna(data[col].mean(), inplace=True)
    
    # Drop rows with any remaining null values
    data.dropna(inplace=True)
    
    # Display the total null values after processing
    total_null_values_after = data.isnull().sum().sum()
    print(f"Total null values after processing: {total_null_values_after}\n")
```
